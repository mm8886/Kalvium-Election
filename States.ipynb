```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://results.eci.gov.in/AcResultGenJune2024/statewiseS188.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles if title.text.strip() and title.text.strip() != 'Status Known For 147 out of 147 Constituencies']
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
for tag in soup.find_all(class_=["tooltip-icon", "tooltip", "tooltiptext", "table", "table-striped"]):
    tag.extract()
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
for row in column_data:
    row_data = row.find_all('td')
    
    seen_values = set()
    individual_row_data = []
    
    for data in row_data:
        text = data.text.strip()
        if text and text not in seen_values:
            individual_row_data.append(text)
            seen_values.add(text)
        if 'Uncontested' in individual_row_data:
            index = individual_row_data.index('Uncontested')
            individual_row_data = individual_row_data[:index] + ['-', '-', '-'] + individual_row_data[index:]
            
    if len(individual_row_data) > 1:
        length = len(df)
        df.loc[length] = individual_row_data
    
```
