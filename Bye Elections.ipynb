```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://results.eci.gov.in/AcResultByeJune2024/ConstituencywiseS25113.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
world_titles = table.find_all('th')
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
world_table_titles = world_table_titles[:7] 
print(world_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data if data.text.strip()]
    
    if len(individual_row_data) == len(df.columns):
        length = len(df)
        df.loc[length] = individual_row_data

total_evms = df['EVM Votes'].astype(int).sum()
total_postal = df['Postal Votes'].astype(int).sum()
total_votes = df['Total Votes'].astype(int).sum()

bottom_row_data = {
    'S.N.': ' ',
    'Candidate': 'Total',
    'Party': ' ',  
    'EVM Votes': total_evms,
    'Postal Votes': total_postal,        
    'Total Votes': total_votes,
    '% of Votes': ' '
}

bottom_row_df = pd.DataFrame([bottom_row_data])
df = pd.concat([df, bottom_row_df], ignore_index=True)
```
